# Carga de datos de sensores a Azure Functions

Este notebook envía datos desde un CSV al endpoint de ingestión.

```python
import pandas as pd
import requests

API_PUT_URL = "http://localhost:7071/api/sensors"
API_PREDICT_URL = "http://localhost:7071/api/predictions"
CSV_FILE = "sensor_data_april_2026.csv"
```

```python
df = pd.read_csv(CSV_FILE)
df.head()
```

```python
for i, row in df.iterrows():
    payload = {
        "machine_id": row["machine_id"],
        "timestamp": row["timestamp"],
        "variable": row["variable"],
        "value": float(row["value"])
    }

    response = requests.put(API_PUT_URL, json=payload)
    print(response.status_code)
    if i % 100 == 0:
        print(f"Sent {i} records, status: {response.status_code}")
```

## Verificación

Después de ejecutar, revisa Azurite o consulta los endpoints de predicción.

```python
# Obtengo las predicciones de todas las maquinas

response = requests.get(API_PREDICT_URL, json=payload)
print(response.text)
```

```python
# Obtengo las predicciones de una maquina paticular

MACHINE = "PUMP-0001"
API_PREDICT_MACHINE_URL = f"http://localhost:7071/api/machines/{MACHINE}/prediction"

response = requests.get(API_PREDICT_MACHINE_URL, json=payload)
print(response.text)
```
